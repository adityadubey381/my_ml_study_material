```python

import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
     
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python

rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

Bagging is a method for training multiple decision trees, while random forest is an ensemble of decision trees that uses randomness to improve predictions.

Explanation:-

Bagging, or bootstrap aggregating, is a common ensemble method that uses random sampling to create a forest of decision trees. Random forest is an extension of bagging that uses randomness to improve variable selection. 
The goal of both bagging and random forest is to combine multiple models to create a more powerful and robust model. 

Ensemble methods:-

Ensemble methods are a collection of models that are combined to make predictions. Other ensemble methods include boosted trees and stacking.
